## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Gender Column
print(df['Gender'].unique())
newGender= LabelEncoder()
df['Gender']= newGender.fit_transform(df['Gender'])
df['Gender'].unique()
```

```python
# Married Column
print(df['Married'].unique())
newMarried= LabelEncoder()
df['Married']= newMarried.fit_transform(df['Married'])
df['Married'].unique()
```

```python
# Property_Area Column
print(df['Property_Area'].unique())
newProperty_Area= LabelEncoder()
df['Property_Area']= newProperty_Area.fit_transform(df['Property_Area'])
df['Property_Area'].unique()
```

```python
# Education Column
print(df['Education'].unique())
newEducation= LabelEncoder()
df['Education']= newEducation.fit_transform(df['Education'])
df['Education'].unique()
```

```python
# Self_Employed
print(df['Self_Employed'].unique())
newSelf_Employed= LabelEncoder()
df['Self_Employed']= newSelf_Employed.fit_transform(df['Self_Employed'])
df['Self_Employed'].unique()
```

```python
# Loan_Status

print(df['Loan_Status'].unique())
newLoan_Status= LabelEncoder()
df['Loan_Status']= newLoan_Status.fit_transform(df['Loan_Status'])
df['Loan_Status'].unique()
```

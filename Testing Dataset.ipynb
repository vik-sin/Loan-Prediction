# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Gender Column
print(testdf['Gender'].unique())
newGender= LabelEncoder()
testdf['Gender']= newGender.fit_transform(testdf['Gender'])
testdf['Gender'].unique()
```

```python
# Married Column
print(testdf['Married'].unique())
newMarried= LabelEncoder()
testdf['Married']= newMarried.fit_transform(testdf['Married'])
testdf['Married'].unique()
```

```python
# Property_Area Column
print(testdf['Property_Area'].unique())
newProperty_Area= LabelEncoder()
testdf['Property_Area']= newProperty_Area.fit_transform(testdf['Property_Area'])
testdf['Property_Area'].unique()
```

```python
# Education Column
print(testdf['Education'].unique())
newEducation= LabelEncoder()
testdf['Education']= newEducation.fit_transform(testdf['Education'])
testdf['Education'].unique()
```

```python
# Self_Employed
print(testdf['Self_Employed'].unique())
newSelf_Employed= LabelEncoder()
testdf['Self_Employed']= newSelf_Employed.fit_transform(testdf['Self_Employed'])
testdf['Self_Employed'].unique()
```

```python
# Loan_Status

print(testdf['Loan_Status'].unique())
newLoan_Status= LabelEncoder()
testdf['Loan_Status']= newLoan_Status.fit_transform(testdf['Loan_Status'])
testdf['Loan_Status'].unique()
```
